```python
# 1) Importez les librairies qu’il nous faut (pandas, seaborn, matplotlib, numpy)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mat
```

```python
df = pd.read_csv('./data/house_pricing.csv')
```

```python
# 2) Donnez la liste des variables présentes dans ce dataset ainsi que leur nature (sont elles
# qualitatives, quantitatives, discrète etc...) et leur type (float, int, str etc...)
df.info()
```

```python
# 3) En créant un nuage de points, regardez comment se comporte la colonne LotArea par
# rapport au SalesPrices
sns.relplot(x='LotArea', y='SalePrice', data=df)
```

```python
# 4) Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea
# inférieur à 20 000 pieds carrés et un prix inférieur à 500 000$

# dataMaison = df[ (df['LotArea'] < 20000) & (df['SalePrice'] < 500000)]
dataMaison = df.query("LotArea < 20000 & SalePrice < 500000")
sns.relplot(x='LotArea', y='SalePrice', data=dataMaison)
```

```python
# 5) En créant un nuage de points, regardez la relation
# entre le LotFrontage et le LotArea

sns.relplot(x='LotFrontage', y='LotArea', data=df)
```

```python
# 6) De la même manière, affinez votre visualisation en ne gardant 
# uniquement les maisons qui ont un LotFrontage inférieur à 200 pieds carrés
# et un LotArea inférieur à 100000 pieds carré

dataMaison = df.query("LotFrontage < 200 & LotArea < 100000")
sns.relplot(x='LotFrontage', y='LotArea', data=dataMaison)
```
